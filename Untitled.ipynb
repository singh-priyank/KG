```python
from nltk.corpus import wordnet

# Function to predict edges 
def predict_edge(G,G1,G2):
    
    g1 = [node for node in G1.nodes(data=True)]
    g2 = [node for node in G2.nodes(data=True)]
    
    
    for s in g1:
        cb = wordnet.synsets(s[1]['labelV'])
        if len(cb)==0:
            continue
        cb = cb[0]
        for e in g2:
            ib = wordnet.synsets(e[1]['labelV'])
            if len(ib)==0:
                continue
            ib = ib[0]
            #print(s[1]['labelV'],e[1]['labelV'],ib.wup_similarity(cb))
            # Condition if similarity is more than 50% and path is already not there 
            if ib.wup_similarity(cb)>=0.5:
                G.add_edge(s[0],e[0])
                tt='........'
                print(s[1]['labelV'],e[1]['labelV'],tt,s[0],e[0])
                print(ib.wup_similarity(cb))

if __name__=="__main__":
    #predict_edge(G2)
    #predict_edge(G1)
    # Merging both the graphs
    F = nx.compose(G2,G1)
    predict_edge(F,G1,G2)
```

```python
#Print final graph 

nx.draw(F,with_labels=True)
plt.show()
for node in F.nodes(data=True):
    print(node)
```

```python
#Find subgraph with depth = radius
ego_graph = nx.ego_graph(F, '0', radius=3)

# plot to check
nx.draw(ego_graph,with_labels=True); plt.show()
```
